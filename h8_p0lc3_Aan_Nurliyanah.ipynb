## Penjabaran Masalah


Soal 1 : Tampilkan tabel crime sampai 5 baris pertama

Soal 2 : Apa saja jenis tindak kejahatan yang terjadi di kota Austin selama tahun 2016?

Untuk mencari jenis tindak kejahatan apa saja menggunakan kolom primary_type dan description. Kemudian di cari berdasarkan klausa where dimana data yang diminta hanya tahun 2016 saja diambil dari kolom year, data diurutkan secara ascending dari berdasarkan primary_type.

```python
jenis_kejahatan = client.query('''
SELECT primary_type, description
FROM `bigquery-public-data.austin_crime.crime` 
WHERE year = 2016
ORDER BY primary_type ASC
''').to_dataframe()

jenis_kejahatan
```

Dari hasil pencarian diatas didaptkan jenis jenis type kejahatan beserta deskripsinya diantaranya ada Agg Assault, dan Theft. 

Soal 3 : Berapa banyak kasus tindak kejahatan berdasarkan kasus pembersihannya (clearance status) selama tahun 2016?

Untuk mencari banyaknya kasus tindak kejahatan kita cari menggunakan fungsi count, yang dihitung adalah dari deskripsinya kemudian dicari menggunakan klausa where dimana data yg diminta di tahun 2016 saja di kolom year, kemudian data di group by berdasarkan clearance statusnya. Untuk total case diurutkan secara descending dari besar ke kecil.

```python
kasus_kejahatan = client.query('''
SELECT 
  COUNT (description) AS total_case,
  clearance_status
FROM `bigquery-public-data.austin_crime.crime` 
WHERE year = 2016
GROUP BY clearance_status
ORDER BY total_case DESC
''').to_dataframe()

kasus_kejahatan
```

Dari perhitungan diatas didapatkan total case dengan cleareance status 'not cleared' adalah yang tertinggi yaitu 29715, artinya banyak kasus yang tidak terpecahkan di tahun 2016 ini. Jumlah ini sangat ekstreme dibandingkan dengan kasus yang sudah cleared/ terselesaikan. 

Saol 4 : Jenis kejahatan apa saja yang paling banyak belum terselesaikan kasusnya di tahun 2016?

Untuk mencari jenis kejahatan yang paling banyak belum terselesaikan dicari dengan menggunakan fungsi count untuk menghitung total case, dihitung berdasarkan deskripsinya, kemudian di group by berdasarkan primary_type, description dan clearence status. Dimana data yang dicari adalah di tahun 2016 dan status celaresnce = 'not cleared'. Data diurutkan berdasarkan total case dari yang terbesar ke kecil.

```python
kasus_belum_selesai = client.query('''
SELECT 
  primary_type, description , clearance_status,
  COUNT (description) AS total_case,
FROM `bigquery-public-data.austin_crime.crime` 
WHERE year = 2016 AND 
  clearance_status = 'Not cleared'
GROUP BY primary_type, description, clearance_status
ORDER BY total_case DESC

''').to_dataframe()

kasus_belum_selesai
```

Dari perhitungan diatas didapatkan hasil yaitu paling banyak total 8607 kasus yang tidak terselesaikan yaitu ada pada jenis kejahatan pencurian (theft). 

Soal 5 : Di bulan apa di tahun 2016 kasus pencurian (theft) sangat banyak terjadi?

Untuk mencari bulan apa di tahun 2016 kasus theft banyak terjadi menggunakan subquery dimana pertama kita mencari data perbulan dahulu dengan menggunakan fungsi extraxt pada timestamp dimana kejadian pencurian dimulai, kemudian di hitung total case nya dengan menggunakan fungsi count, dengan kondisi tahun yg dicari 2016 dan jenis kejahatan = theft. Setelah didapatkan data setiap bulannya kemudian di urutkan dengan menggunakan fungsi row_number dan di partisi per bulan untuk mencari urutan pertama saja yang ditampilkan perbulannya. 

```python
kasus_bulan = client.query('''
SELECT
crime_type, description, month, total_case
FROM (SELECT *, row_number() over (
PARTITION by month 
ORDER BY total_case DESC) AS URUTAN
FROM 
(SELECT 
  primary_type as crime_type,
  description as description,
  EXTRACT (month from timestamp) as month,
  COUNT (description) AS total_case,
FROM `bigquery-public-data.austin_crime.crime` 
WHERE year = 2016 AND 
  primary_type = 'Theft'
GROUP BY month, crime_type, description
ORDER BY total_case DESC))
WHERE URUTAN = 1
ORDER BY total_case DESC
''').to_dataframe()

kasus_bulan
```

Dari perhitungan di atas didapatkan hasil di bulan 12 atau bulan desember terjadi kasus pencurian terbanyak yaitu  925 kasus. 

Soal 6 : Di distrik apa yang paling banyak terjadi kasus pencurian selama 2016?

Untuk mencari district menggunakan metode count di bagian description untuk mencari total case kemudian di group by berdasarkan district .

```python
kasus_district = client.query('''
SELECT 
primary_type as crime_type,
description as description,
district as district,
COUNT (description) AS total_case,
FROM `bigquery-public-data.austin_crime.crime` 
WHERE year = 2016 AND 
primary_type = 'Theft'
GROUP BY crime_type, description, district
ORDER BY total_case DESC
''').to_dataframe()

kasus_district
```

dari hasil perhitungan diatas didaptakan diatrict yang terjadi banyak kejahatan adalah di district D dengan total 1597 kasus.

Soal 7 : Berapa lama rata-rata waktu (dalam hari) kasus tindak kejahatan 'theft' terselesaikan sejak kasus tersebut terangkat selama tahun 2016? (Hint: Hitung selisih hari dari timestamp hingga clearance date)

Untuk menghitung rata rata waktu menggunakan fungsi timestamp_diff dimana kita mencari selisih waktu dalam hari antara clearence date dan timestamp. 
sebelum mencari rata rata saya menampilkan dulu jangka waktu / selisih waktunya secara rinci diurutkan dari besar ke kecil. 

```python
sebelum_rata_rata = client.query('''
SELECT
	clearance_status ,timestamp, clearance_date,
    timestamp_diff(clearance_date,timestamp,day) as Jangka_waktu_hari
 FROM bigquery-public-data.austin_crime.crime
 WHERE year = 2016 AND
 primary_type = 'Theft' AND 
 clearance_status = 'Cleared by Arrest' OR 
 clearance_status = 'Cleared by Exception'
 ORDER BY Jangka_waktu_hari DESC
''').to_dataframe()
sebelum_rata_rata
```

Setelah mendapatkan hasil, kita lanjuta subquery untuk mencari rata rata dengan fungsi avg. 

```python
rata_rata = client.query('''
SELECT
AVG(Jangka_waktu_hari) AS rata_rata,
FROM(SELECT
	clearance_status ,timestamp, clearance_date,
    timestamp_diff(clearance_date,timestamp,day) as Jangka_waktu_hari
 FROM bigquery-public-data.austin_crime.crime
 WHERE year = 2016 AND
 primary_type = 'Theft' AND 
 clearance_status = 'Cleared by Arrest' OR 
 clearance_status = 'Cleared by Exception'
 ORDER BY Jangka_waktu_hari DESC)
''').to_dataframe()

rata_rata
```

dari hasil perhitungan diatas didapatkan rata rata lamanya kasus terselesaikan adalah 25 hari . Untuk yang paling lama adalah 359 hari yang berakhir dengan cleared by arrest.
